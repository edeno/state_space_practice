Variables:
+ $y^i_{kt}$ - spike count of neuron $i$ in time bin $t$ of trial $k$
+ $y_k$ - spike count of all neurons in trial $k$
+ $x_{kt}$ - latent variable of neuron $i$ in time bin $t$ of trial $k$
+ $C$ - $q$ x $p$ matrix of coefficients that map latent variables to spike counts
+ $d$ - mean spike count of each neuron
+ $D$ - history term for each neuron
+ $s_t$ - recent spike counts of all neurons in time bin $t$
+ $Q_kt$ - covariance matrix of latent variables in time bin $t$ of trial $k$
+ $A$ - transition matrix for latent variables
+ $b$ - bias term for latent variables (input driving inputs)
+ $q$ - number of cells
+ $p$ - number of latent variables
+ $T$ - number of time bins
+ $K$ - number of trials


Model:
$$x_{k1} \sim \mathcal{N}(x_0, Q_0)$$
$$x_{k(t+1)}\mid x_{kt} \sim \mathcal{N}(Ax_{kt} + b, Q)$$
$$\lambda^i_{kt} = E[y^i_{kt}] = \exp(Cx_{kt} + Ds_t + d)$$

Learn:
$$\Theta = \{C, D, d, A, Q, Q_0, x_0\}$$

$$p(x_k \mid y_k, \Theta) \approx \mathcal{N}(\mu_k, \Sigma_k)$$

$$ \mu_k = \argmax_{x} p(x \mid y_k, \theta)$$
$$ \Sigma_k = - (\nabla\nabla_x log p(x \mid y_k, \Theta)|_{x=\mu_k})^{-1}$$

$$\log p(x_k \mid y_k, \Theta) =
const + \sum_{t=1}^T \sum_{i=1}^q y^i_t \log \lambda_t - \lambda_t -
\frac{1}{2} \sum_{t=0}^{T-1} (x_{k, t+1} - Ax_{kt} - b)' Q^{-1} (x_{k,t+1} - Ax_{kt} - b_t)$$
